```python
import pandas as pd

data = {
    "Name": ["A","B","C","C","D",None],
    "Marks": [85,90,None,78,88,76],
    "Age": [20,21,19,19,None,20],
    "Date": ["2024-01-01","2024-02-01","wrong","2024-03-01","2024-04-01","2024-05-01"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df = df.fillna(0)
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
print(df)
```

```python
df = df.drop_duplicates()
print(df)
```

```python
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
df.to_csv("cleaned_data.csv", index=False)
print("Cleaned data saved")
```
